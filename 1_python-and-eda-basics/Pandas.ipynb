### More on CSVs

```python
from io import StringIO, BytesIO 
#import numpy as np

csv_data = ('Col1,Col2,Col3\n'
          '1,A,123ABC\n'
          '2,B,12DEF\n'
          '3,C,123GHI')
```

```python
type(csv_data)
```

```python
df1 = pd.read_csv(StringIO(csv_data))
```

```python
df1.head()
```

```python
pd.read_csv(StringIO(csv_data),usecols = ['Col1','Col3'])
```

```python
pd.read_csv(StringIO(csv_data),usecols = ['Col1','Col3']).to_csv('Test1.csv')
```

```python
df1 = pd.read_csv(StringIO(csv_data), dtype=object)
```

```python
df1
```

```python
df1['Col1'][1] #Note its a string not a number
```

```python
df1 = pd.read_csv(StringIO(csv_data), dtype={'Col1':int, 'Col2':object, 'Col3':object })
```

```python
df1
```

```python
df1['Col1'][1] #Notice its a number now
```

```python
str_csv_data = ('Col1, Col2, col3, Col4\n'
                  '4,Apple,100 Cal\n'
                  '8, Bananna, 65 Cal\n')
```

```python
pd.read_csv(StringIO(str_csv_data)) ## Currencly index is 0 and 1
```

```python
pd.read_csv(StringIO(str_csv_data), index_col='Col1') ## Now Col1 is index
```

```python
pd.read_csv(StringIO(str_csv_data), index_col=0) ## Now Col1 is index again
```

```python
str_csv_data1 = ('Col1, Col2, col3\n'
                  '4,Apple,100 Cal,\n'
                  '8, Bananna, 65 Cal,\n')
```

```python
pd.read_csv(StringIO(str_csv_data1)) 
```

```python
pd.read_csv(StringIO(str_csv_data1), index_col=False) ## Notice the diffrence due to index = false
```

```python
pd.read_csv(StringIO(str_csv_data1), usecols=['Col1','Col2','Col3'], index_col=False) ## Need to check
```
