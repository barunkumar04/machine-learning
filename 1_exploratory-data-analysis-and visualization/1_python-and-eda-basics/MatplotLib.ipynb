## MatplotLib tutorial

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='red')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
```

```python
plt.savefig('FirstGraph.png')
```

```python
y = x*x
plt.plot(x,y)
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'r*--')
```

```python
plt.plot(x,y,'r*', linestyle='dashed', linewidth=2, markersize=12)
```

```python
##  Creating subplots

plt.subplot(2,2,1) ## 2 row and 2 columns
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g--')
```

### Box plot - it helps to find percentile

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangular box plot

plt.boxplot(data, vert=True, patch_artist=True);
```

```python
plt.boxplot(data, vert=True, patch_artist=False);
```

### Pie Chart 

```python
sizes = [250,300,215,190]
labelsData = ['Python', 'Java', 'C', 'C++']
colorsData = ['Red', 'Yellow', 'Blue', 'Green']
explodeData = [0.1, 0.0, 0.0, 0.0]

plt.pie(sizes, explode=explodeData, labels=labelsData, colors=colorsData, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
