```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()

# 'tip' is a dependent feature
# total_bill, sex, smoker, day etc are independent feature
```

### Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matric (table) between two discreat dimension or event types. It is very importantn in Feature Selection.

```python
df.corr() # it will show correlation for integer/flaot only
```

```python
df.dtypes
```

```python
# lets visualize the correlation

sns.heatmap(df.corr())
```

### Join Plot

A Join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexpin plot, a 2D histogram or 2D density plot.

#### Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # Reggression line
```

### Pair plot

A 'pair plot' is also known as a scatterplot, in which one variable in the same data row, is matched with another variable's value. Pair plots are just elaborations on this, showing all variables paired with all the other variables. 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

### Dist plot

Dist plot helps to check the distributions of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)

# kde= - Kernal Density Estimation
# ~80 people given ~2 or ~3$ tip
# When kde is disabled then only it will show count otherwise it will represent %
```

### Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this session following will be covered

1. Box plot
2. Violin Plot
3. Count Plot
4. Bar Plot

```python
## Count Plot

sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df, )
```

```python
sns.countplot(y='day', data=df, )
```

```python
## Bar plot

# In count plot one axis is always count, but in Bar plot we must give 'Y' axis

sns.barplot(y='total_bill', x='smoker', data=df)
```

```python
sns.barplot(y='total_bill', x='day', data=df)

# more billing was done on sunday
```

### Box plot
A box and whisker plot(i.e. boxplot) is a graph that presents information from a five-number summary.

Five-numbers are:
1. the bottom line - 0 persentile
2. lower line of box - 25 percentile
3. line inside the box = 50 percentile
4. upper line of box - 75 percentile
5. the top line - 100 percentile

6. dots are outliers

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot('day','total_bill', data=df)
```

```python
sns.boxplot('day','total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day', hue='smoker', data=df)
```

### Violin Plot

Violin plot helps us to see both the distribution of data in terms of (kde)Kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```

```python
irisDF = sns.load_dataset('iris')
```

```python
# Lets observe the data and type

type(irisDF) # So its a DataFrame
```

```python
irisDF.head()
```

```python
irisDF.corr() # note Correlation is an attribute of dataframe, not Seaborn

## Observations out of this correlation
# When sepal length increase, sepal width decrease (the -ve number -0.117570)
# When sepal length increase, petal length increase
# When sepal length increase, petal width increase
```

```python
# check heatmap

sns.heatmap(irisDF.corr()) 

# Note here, you need to provide correlation, but the entire dataframe
```

#### A. non-Categorical plots

```python
# JOIN plot

sns.jointplot(x='petal_length', y='petal_width', data=irisDF, kind='hex')

## Observation
# When petal length increase, petal width also increses
# When petal length is 4-6, mejority of petal width is between 1 - 2 (i guess in CM)
```

```python
# Changing 'hex' to have distribution
sns.jointplot(x='petal_length', y='petal_width', data=irisDF, kind='reg')
```

```python
# PAIR plot

sns.pairplot(irisDF)

## Observation
# Same as pair plot, but it shows in scattered way
```

```python
sns.pairplot(irisDF, hue = 'species')

## Observation - It addes species information.
```

```python
## Dist plot

sns.distplot(irisDF['petal_length'])
```

```python
sns.distplot(irisDF['petal_length'], kde=False)
```

#### B. Categorical Plots

1. Count Plot - where one axis is count
2. Bar plot - where user need to specify X and Y axis
3. Box plot - Where 5 points can be understood
4. Violin Plot - Combination of Box plot and distibution

```python
# COUNT plot

sns.countplot('species', data=irisDF, palette='rainbow')

```

```python
# BAR Plot

sns.barplot(x='species',y='petal_width', data=irisDF)
```

```python
sns.barplot(x='species',y='petal_width', data=irisDF, dodge=True)
```

```python
## BOX plot

sns.boxplot(x='species',y='petal_width', data=irisDF)
```

```python
## VIOLIN plot

sns.violinplot(x='species',y='petal_width', data=irisDF)
```
